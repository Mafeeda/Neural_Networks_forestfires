### Since the number of columns are more,let's use PCA (PCA - used for Dimensionality Reduction)

```python
pca = PCA(n_components = 28)
pca_values = pca.fit_transform(data_norm)
pca_values
```

```python
var = pca.explained_variance_ratio_
var
```

```python
cum = np.cumsum(np.round(var,decimals = 4)*100)
cum
```

```python
plt.figure(figsize = (15,5))
plt.plot(cum, color = 'red')
```

```python
final = pd.concat([pd.DataFrame(pca_values[:,0:22],columns = ['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10','pc11',
                                                            'pc12','pc13','pc14','pc15','pc16','pc17','pc18','pc19','pc20',
                                                            'pc21','pc22']), forestfire_data[['size_category']]], axis = 1)
final.size_category.replace(('large','small'),(1,0),inplace = True)
final
```

```python
#split the data into x and y

array = final.values
x = array[:,0:22]
y = array[:,22]
```

## Model Building

```python
model = Sequential()
model.add(Dense(40, input_dim = 22, activation = 'relu'))
model.add(Dense(11, activation = 'relu'))
model.add(Dense(1, activation ='sigmoid'))
```

```python
# Compile the model

model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
# Fit the model with 30% of test data

model_fit = model.fit(x, y, validation_split = 0.3, epochs = 100, batch_size = 10)
```

```python
# Evaluate the model

scores = model.evaluate(x, y)

print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
model.predict(x)
```
